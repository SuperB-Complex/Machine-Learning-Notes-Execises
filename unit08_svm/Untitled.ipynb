```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1.3, 2.1, 2.8, 4.2, 5.7])
def getFunctionJ(z):
    interval = np.arange(0, z.shape[0], 1)
    result = 0
    recordviolates = []
    recordmisclassify = []
    for i, v in enumerate(interval):
        if z[i] > 0:
            y = 1
        else:
            y = -1
        temp = 1 - y * z[i]
        if (temp  > 0) and (temp <= 1) :
            recordviolates.append(i)
        if temp > 1:
            recordmisclassify.append(i)
        result += np.amax(np.array([0, temp]))
    return result, recordviolates, recordmisclassify
t = np.random.uniform(0, 5, 100).reshape(-1, 1)
z = x[None, :] - t[:]
J = np.zeros((t.shape[0], 1))
RV = []
RM = []
for i, v in enumerate(t):
    tempv = []
    tempm = []
    J[i], tempv, tempm = getFunctionJ(z[i])
    RV.append(tempv)
    RM.append(tempm)
plt.scatter(t, J)
mint = np.argmin(J)
print(mint)
print(t[mint])
print(J[mint])
print(RV[mint])
print(RM[mint])
```
